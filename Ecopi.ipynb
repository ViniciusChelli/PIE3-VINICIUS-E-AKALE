```python
from flask import Flask, jsonify
import random
import datetime

app = Flask(__name__)

@app.route("/data")
def get_data():
    data = {
        "temperatura": round(random.uniform(20, 35), 2),
        "umidade": round(random.uniform(40, 90), 2),
        "timestamp": datetime.datetime.now().strftime("%H:%M:%S")
    }
    return jsonify(data)

if __name__ == "__main__":
    app.run(debug=True)
```
